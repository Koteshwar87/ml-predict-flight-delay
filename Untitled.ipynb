```python
def read_airports_data():
    airports = pd.read_csv('airports.csv')
    print(airports.columns)
```

```python
def read_flights_data():
    flights = pd.read_csv('flights.csv', low_memory=False)
    # print(flights.columns)
    # print(flights.head())
    #flights_seg = flights[0:50000]
    flights_seg = flights.sample(n=50000, axis=0, random_state=2)
    return flights_seg
```
